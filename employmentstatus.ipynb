```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import geopandas as gpd
import folium
import matplotlib.ticker as ticker
import matplotlib.cm as cm
import numpy as np
import csv
from tabulate import tabulate
```

```python
columns_to_keep = [0,1,10,11]

df = pd.read_csv('employmentstatus19.csv',usecols=columns_to_keep, skiprows=1)
new_column_names = {'Unnamed: 0': 'code', 'Unnamed: 1': 'area', 'number.2':'self_employed', 
                   'denominator.2':'total_employed'}
df.rename(columns=new_column_names, inplace=True)
df
```

```python
year = 2019  
df.insert(0, 'year', year)
df
```

```python
condition = df['area'].isin(['Inner London', 'Outer London', 'North East','North West',
                            'Yorkshire and The Humber','East Midlands','West Midlands','East','London',
                            'South East','South West','England','Wales','Scotland',
                            'Northern Ireland','England and Wales','Great Britain','United Kingdom'])

DF = df[~condition]
DF
```

```python


for col in DF.columns[3:]:
    # Convert the column to string in case it's not
    DF[col] = DF[col].astype(str)
    # Now you can use str accessor to replace commas
    DF[col] = DF[col].str.replace(',', '')
    # Convert the column to numeric
    DF[col] = pd.to_numeric(DF[col], errors='coerce')

DF.fillna(value=np.nan, inplace=True)

DF.iat[1, 3] = 0
# Drop rows with missing data
df1 = DF.dropna()
```

```python
df1.to_csv('selfemp19.csv',index = False)
```

```python
df2 = pd.read_csv('employmentstatus20.csv',usecols=columns_to_keep, skiprows=1)

df2.rename(columns=new_column_names, inplace=True)

year2 = 2020  
df2.insert(0, 'year', year2)

DF2 = df2[~condition]
for col in DF2.columns[3:]:
    DF2[col][1:] = pd.to_numeric(DF2[col][1:].str.replace(',', ''), errors='coerce')

DF2.iat[1, 3] = 0
df22 = DF2.dropna()
df22.to_csv('selfemp20.csv',index = False)
df22
```

```python
df3 = pd.read_csv('employmentstatus21.csv',usecols=columns_to_keep, skiprows=1)

df3.rename(columns=new_column_names, inplace=True)

year3 = 2021  
df3.insert(0, 'year', year3)

DF3 = df3[~condition]
for col in DF3.columns[3:]:
    DF3[col][1:] = pd.to_numeric(DF3[col][1:].str.replace(',', ''), errors='coerce')

DF3.iat[1, 3] = 0
df33 = DF3.dropna()
df33.to_csv('selfemp21.csv',index = False)
df33
```

```python
files = ['selfemp19.csv', 'selfemp20.csv','selfemp21.csv']
df_list = []

for file in files:
    df = pd.read_csv(file)  
    df_list.append(df)

combined_df = pd.concat(df_list, ignore_index=True)

combined_df.to_csv('selfemployed_cleaned.csv', index=False)
```
