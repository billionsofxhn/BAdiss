```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import geopandas as gpd
import folium
import matplotlib.ticker as ticker
import matplotlib.cm as cm
import numpy as np
import csv
from tabulate import tabulate
```

```python

files = ['PBC19.csv','PBC20.csv','PBC21.csv','PBC22_now.csv' ]


years = ['2019', '2020', '2021', '2022/23']


df_list = []


for file, year in zip(files, years):
    df = pd.read_csv(file)
    df.insert(0, 'Year', year)
    df_list.append(df)


combined_df = pd.concat(df_list, ignore_index=True)
combined_df = combined_df.applymap(lambda x: x.lower() if isinstance(x, str) else x)
combined_df.rename(columns=lambda x: x.lower(), inplace=True)
combined_df.replace('', np.nan, inplace=True)
```
